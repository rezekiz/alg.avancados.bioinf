```python
from grafos import WeightedGraph

grafop = WeightedGraph()
grafop.add_edges(['1 -> 2, 3, 4','2 -> 5, 6', '5 -> 7', '3 -> 6, 8', '4 -> 8' ])
grafop.add_weight('1','2',14)
grafop.add_weight('4','8',4)
grafop._build()
grafop.show()
```
